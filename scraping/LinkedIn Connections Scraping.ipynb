```python
time.sleep(3)
for i in tqdm(range(1, 101)):
    url = "https://www.linkedin.com/search/results/people/?connectionOf=%5B%22ACoAAAFEuIQBT96iT1CvOPY12oImzz32aplULZQ%22%5D&network=%5B%22S%22%5D&origin=FACETED_SEARCH&page="+str(i)+"&sid=ghh"
    driver.get(url)	 # this will open the link
    time.sleep(2)
    src = driver.page_source
    soup = BeautifulSoup(src, 'html.parser')
    for i in soup.find_all("li", {"class":"reusable-search__result-container"}):
        conns.add(i.find("a", href=True)['href'].split('?')[0])
```

```python
len(conns)
```

```python
data = []
for url in tqdm(conns):
    
    driver.get(url)	 # this will open the link
    time.sleep(2)
    src = driver.page_source
    soup = BeautifulSoup(src, 'html.parser')
    try:
        name = soup.find("h1", {"class": "text-heading-xlarge inline t-24 v-align-middle break-words"}).text
    except:
        name = ""
    try:
        #experience = soup.find_all("span", {"class": "t-bold mr1 hoverable-link-text"})#.find('ul')
        if soup.find("span", {'class':"pvs-entity__path-node"}) is None:
            company = soup.find("span", {"class":"t-14 t-normal"}).span.text.split(" · ")[0]
        else:
            company = soup.find_all("span", {"class": "t-bold mr1 hoverable-link-text"})[0].span.text
        #result = re.search('<!-- -->(.*)<!-- -->', str(company))
        #company = result.group(1)
    except:
        company = ""

    try:
        if soup.find("span", {'class':"pvs-entity__path-node"}) is None:
            role = soup.find("span", {"class":"t-bold mr1"}).span.text#experience[0].span#experience[1].span
        else:
            role = soup.find_all("span", {"class": "t-bold mr1 hoverable-link-text"})[1].span.text
        #result = re.search('<!-- -->(.*)<!-- -->', str(role))
        #role = result.group(1)
    except:
        role = ""
        
        url += "/overlay/contact-info/"
        driver.get(url)
        time.sleep(2)
        src = driver.page_source
        soup = BeautifulSoup(src, 'html.parser')
    try:
        email = soup.find("section", {"class":"pv-contact-info__contact-type ci-email"}).find("a").text.strip()
    except:
        email = ""
    
    try:
        phone = soup.find("section", {"class":"pv-contact-info__contact-type ci-phone"}).find("span", {"class":"t-14 t-black t-normal"}).text.strip()
    except:
        phone = ""
    
    data.append([name, company, role, url, email, phone])
    name, company, role, email, phone = "", "", "", "", ""
    
```

```python
df = pd.DataFrame(data, columns = ['Name', 'Company', 'Position', 'LinkedIn', 'Email', 'Phone'])
```
